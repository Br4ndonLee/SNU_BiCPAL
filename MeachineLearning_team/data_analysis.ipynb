```python
# csv 파일 로드
PATH = "AAOS.csv"
df = pd.read_csv(PATH, encoding='cp949')
df.shape
print(df.head())
print(df.columns)
```

```python
# 데이터를 로드해 오면 object 타입입니다.
df["일시"].head(2)
```

```python
# datetime 형식으로 변경해 줍니다.
df["일시"] = pd.to_datetime(df["일시"])
df["일시"].head(1)
```

```python
# 연, 월, 일 파생변수를 생성합니다.
df["연"] = df["일시"].dt.year
df["월"] = df["일시"].dt.month
df["일"] = df["일시"].dt.day
df[["일시", "연", "월", "일"]].head()
```

```python
plt.figure(figsize=(15, 4))
plt.title("일자별 1.5m 평균 기온")
sns.lineplot(data=df, x="일시", y="1.5M 평균 기온(°C)", ci=None)
```

```python
plt.figure(figsize=(15, 4))
plt.title("일자별 1.5m 평균 습도")
sns.lineplot(data=df, x="일시", y="1.5M 평균 습도(%)", ci=None)
```
