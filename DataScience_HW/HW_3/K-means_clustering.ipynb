```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# (a) Plot the observations
X = np.array([[1, 4], [1, 3], [0, 4], [5, 1], [6, 2], [4, 0]])

plt.scatter(X[:, 0], X[:, 1])
plt.title("Observations")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

```python
# (b) Randomly assign a cluster label to each observation
kmeans = KMeans(n_clusters=2, init='random', max_iter=100, n_init=1, random_state=0)
y_kmeans = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_

print("Cluster labels for each observation: ", y_kmeans)
print("Centroids for each cluster: ", centroids)
```

```python
# (c)
while True:
    old_centroids = centroids.copy()
    y_kmeans = kmeans.predict(X)
    centroids = kmeans.cluster_centers_
    if np.all(old_centroids == centroids):
        break

print("Cluster labels for each observation: ", y_kmeans)
```

```python
# (d)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=300, c='red')
plt.title("Observations with Cluster Labels")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

(e)
K-Mean clustering 알고리즘에서 얻은 결과는 클러스터 레이블의 초기 random assignment에 따라 달라질 수 있다. 따라서 (b)부터 (d)까지 과정을 여러번 반복하면 클러스터 레이블과 중심이 달라질 수 있다. 하지만 최종 클러스터 레이블과 중심은 몇 번의 반복 후 수렴되어야 한다.
